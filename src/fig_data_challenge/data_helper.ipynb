```python
import pandas as pd
import typo
```

```python
def drop_unnamed_columns(input_df):
    """ drop unnamed columns from dataframe """
    df_ = input_df
    df_ = df_.loc[:, ~df_.columns.str.contains('Unnamed')]
    return df_
```

```python
def drop_rows_empty(
        input_df:object,
        column_list:list
        ):
    """ drop rows where value are None or Empty strings for given list of columns"""
    for col in column_list:
        input_df=input_df[input_df[col].notna()]
        input_df=input_df[input_df[col]!='']
    return input_df
```

```python
# read data
restaurant_df = pd.read_excel('../../data/restaurant_data.xlsx')
restaurant_df = drop_unnamed_columns(restaurant_df)
```
